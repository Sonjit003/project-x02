```python
students = {
    'uid-1': {"name": "John", "surname": "Doe", "email": "john@gmail.com"},
    'uid-2': {"name": "Jane", "surname": "Doe", "email": "jane@gmail.com"},
    'uid-3': {"name": "John", "surname": "Smith", "email": None},
    'uid-4': {"name": "Jane", "surname": "Smith", "email": "jane@gmail.com"},
}

students_attendance = {
    'uid-1': [True, True, True, True, True, True, True, True, False, True],
    'uid-2': [False, False, False, True, True, False, True, True, False, True],
    'uid-3': [True, True, False, False, True, True, True, True, True, True],
    'uid-4': [False, False, False, False, True, False, True, True, False, True],
}

students_grades = {
    'uid-1': {"hw1": 10, "hw2": 8, "hw3": 9},
    'uid-2': {"hw1": 9, "hw2": 8, "hw3": 10},
    'uid-3': {"hw1": 8, "hw2": 9, "hw3": 10},
    'uid-4': {"hw1": 10, "hw2": 8, "hw3": 10},
}
def calculate_avarage(uid:str, data: dict)-> float:
    grades = data[uid]
    grades_values = list(grades.values())
    students_average_grades = sum(grades_values)/len(grades_values)

    return students_average_grades

def define_the_name (uid:str, data:dict)-> str:
    student_info = data[uid]
    student_name = student_info["name"]
    student_surname = student_info["surname"]
    students_full_name = f"{student_name} {student_surname}"
    return students_full_name


def calculate_average_grades(personal_info, students_result):
    result = {}
    for uid in students_result.keys():
        name_score = {}
        name_score["name"] = define_the_name(uid, personal_info)
        name_score["score"] = calculate_avarage(uid, students_result)
        result[uid] = name_score
    return result
p = calculate_average_grades(students, students_grades)
print(p)
```

```python
def calculate_student_atendance (uid: str, data:dict):
    attendance_value = data[uid]
    attendance_percentage = (sum(attendance_value)/len(attendance_value))*100
    return attendance_percentage

def calculate_attendance_percentage ( students_info, st_attendance ):
    result = {}
    for uid in st_attendance.keys():
        per = {}
        per["name"] = define_the_name(uid, students)
        per["per_attendance"] = calculate_student_atendance(uid, st_attendance)
        
        result[uid] = per
    return result


p = calculate_attendance_percentage(students, students_attendance)
print(p)
```

```python
def calculate_students_summary(personal_info:dict, students_results:dict, students_participation:dict)-> dict:
    result = {}
    for uid in students_results.keys():
        summary = {}
        summary["name"] = define_the_name(uid, personal_info)
        summary["score"] = calculate_avarage(uid, students_grades)
        summary["attandence percentage"] = calculate_student_atendance(uid,students_attendance)
        result[uid] = summary
    return result

p = calculate_students_summary(students, students_grades, students_attendance)
print(p)
```

```python
def calculat_student_overall_performance(student_info, st_grades, st_attendance):
    overall_performance = {}
    for uid in st_grades.keys():
        performance = {}
        performance["name"] = define_the_name(uid, student_info)
        performance["overall_score"] = calculate_avarage(uid, st_grades)*0.7 + calculate_student_atendance(uid, students_attendance)*0.3
        overall_performance[uid] = performance
    return overall_performance

p = calculat_student_overall_performance(students, students_grades, students_attendance)

print(p)
```

```python
def calculate_students_rank(st_info, st_grades, st_attendance):

    list_of_overall_performance = [(key, value["overall_score"]) for key, value in calculat_student_overall_performance(students,students_grades,students_attendance).items()]
    sorted_list = sorted(list_of_overall_performance, key= lambda x:x[1], reverse=True)

    student_rank = {}
    rank_counter = 1
    for uid, score in sorted_list:
        student_rank[uid] = {"score": score, "rank": rank_counter}
        rank_counter = rank_counter+1
    return student_rank
    
p = calculate_students_rank(students, students_grades, students_attendance)

print(p)
```
